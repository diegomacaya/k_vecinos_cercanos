#### Paso 1: Carga del conjunto de datos

```python
import pandas as pd
movies_df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv")
credits_df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv")
```

#### Paso 2: CreaciÃ³n de una base de datos

```python
import sqlite3

#Crear base de datos en sql
conexion_movies= sqlite3.connect("../data/raw/movies_db.db")
```

```python
#Transformar los dataframe en formato sql
movies_df.to_sql('Tabla_movies',conexion_movies,if_exists='replace',index=False)
credits_df.to_sql('Tabla_credits',conexion_movies,if_exists='replace',index=False)
```

```python
print(movies_df.columns)
print(credits_df.columns)
```

```python
#Crear una sola tabla

query_movies = """
    SELECT tc.movie_id, tm.title, tm.overview, tm.genres, tm.keywords, tc.cast, tc.crew
    FROM Tabla_movies as tm
    INNER JOIN Tabla_credits as tc
    ON tm.title = tc.title;
"""

df_movies = pd.read_sql_query(query_movies,conexion_movies)
conexion_movies.close()

df_movies.head()
```

#### Paso 3: Transforma los datos

```python
import json 
#selecciona el atributo `name` y reemplaza las columnas `genres` y `keywords`
df_movies['genres'] = df_movies['genres'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
df_movies['keywords'] = df_movies['keywords'].apply(lambda x: [item['name'] for item in json.loads(x)] if pd.notna(x) else None)
```

```python
#Para la columna `cast`, selecciona los tres primeros nombres
df_movies['cast'] = df_movies['cast'].apply(lambda x: [item['name'] for item in json.loads(x)[:3]] if pd.notna(x) else None)
```

```python
#para la columna crew, dejar solo el nombre del director, observaciÃ³n: lambda x: " ".join([crew_member['name'] en este caso, las "",join se utilizan para eliminar los parentecis cuadrados, asociado al nombre del director de la pelÃ­cula
df_movies["crew"] = df_movies["crew"].apply(lambda x: ([crew_member['name'] for crew_member in json.loads(x) if crew_member['job'] == 'Director']))
```

```python
#la columna overview debe quedar con formato de lista
df_movies["overview"] = df_movies["overview"].apply(lambda x: [x])
```

```python
#quitaremos los espacios entre las palabras. 
#Aplica esta funciÃ³n a las columnas `genres`, `cast`, `crew` y `keywords`
df_movies["genres"] = df_movies["genres"].apply(lambda x: str(x).replace(" ","").replace("[","").replace("]",""))
df_movies["cast"] = df_movies["cast"].apply(lambda x: str(x).replace(" ","").replace("[","").replace("]",""))
df_movies["crew"] = df_movies["crew"].apply(lambda x: str(x).replace(" ","").replace("[","").replace("]",""))
df_movies["keywords"] = df_movies["keywords"].apply(lambda x: str(x).replace(" ","").replace("[","").replace("]",""))
df_movies["overview"] = df_movies["overview"].apply(lambda x: str(x).replace("[","").replace("]",""))
```

```python
df_movies.head()
```

Reduciremos nuestro conjunto de datos combinando todas nuestras columnas convertidas anteriores en una sola columna llamada `tags` (que crearemos). Esta columna ahora tendrÃ¡ todos los elementos separados por comas y luego las reemplazaremos por espacios en blanco.

```python
#transformar a string los campos
df_movies["movie_id"] = df_movies["movie_id"].apply(lambda x: [str(x)])
df_movies["title"] = df_movies["title"].apply(lambda x: [str(x)])
df_movies["overview"] = df_movies["overview"].apply(lambda x: [str(x)])
df_movies["genres"] = df_movies["genres"].apply(lambda x: [str(x)])
df_movies["keywords"] = df_movies["keywords"].apply(lambda x: [str(x)])
df_movies["cast"] = df_movies["cast"].apply(lambda x: [str(x)])
df_movies["crew"] = df_movies["crew"].apply(lambda x: [str(x)])
df_movies["tags"] = df_movies["movie_id"] + df_movies["title"] + df_movies["overview"] + df_movies["genres"]+ df_movies["keywords"] + df_movies["cast"]+df_movies["crew"]
df_movies['tags'] = df_movies['tags'].apply(lambda x: ",".join(x).replace(","," "))
```

#### Paso 4: Construye un KNN

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_movies['tags'])

model = NearestNeighbors(n_neighbors=6,algorithm='brute',metric='cosine')
model.fit(tfidf_matrix)

def obtener_recomendaciones(titulo_movie):
    indice_pelicula = df_movies[df_movies['title']==titulo_movie].index[0]
    distancias, indices = model.kneighbors(tfidf_matrix[indice_pelicula])
    peliculas_parecidas = [(df_movies['title'][i],distancias[0][j]) for j,i in enumerate(indices[0])]
    return peliculas_parecidas[1:]

pelicula_consultar = "How to Train Your Dragon"
recomendaciones = obtener_recomendaciones(pelicula_consultar)
for pelicula, distancia in recomendaciones:
    print("-Pelicula {}".format(pelicula))
```
